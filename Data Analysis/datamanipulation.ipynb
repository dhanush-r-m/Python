```python
df = df.rename(columns={'Sales Person':'Sales'})
```

```python
df.head()
```

```python
## create a new column 'Value' with some initial values
df['Value'] = np.random.randint(1, 100, df.shape[0])

## change datatypes
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
```

```python
df.head()
```

```python
df['New Value'] = df['Value'].apply(lambda x: x*2)
```

```python
df.head()
```

```python
## Data Aggregation and grouping
grouped_mean = df.groupby('Product')['Value'].mean()
```

```python
print(grouped_mean)
```

```python
## aggregation with multiple columns
grouped_agg = df.groupby('Product').agg({'Value': 'mean', 'Sales': 'count'})
```

```python
print(grouped_agg)
```

```python
## Merging and Joining
df1 = pd.DataFrame({'Key': ['A0', 'A1', 'A2', 'A3'],'value': [1, 2, 3, 4]})
df2 = pd.DataFrame({'Key': ['B0', 'B1', 'B2', 'B3'],'value': [5, 6, 7, 8]})   
result = pd.concat([df1, df2], axis=1)
                    
```

```python
result
```

```python
pd.merge(df1, df2, on='Key', how='outer')
```

```python
pd.merge(df1, df2, on='Key', how='left')
```

```python
pd.merge(df1, df2, on='Key', how='right')
```
